```python
# import cv2
import numpy as np
import os
from os.path import isfile, join, isdir

# Import own lib
from createVideo import create_video

# ROOT FOLDER
read_root = 'F:\\'
write_root = 'H:\CONVERT_VIDEOS'
```

```python
# Create test list
test_num = np.arange(10,66)
tests = ["{:04d}".format(number) for number in test_num]
tests
```

```python
# tests = os.listdir(root)
for test in tests:
    print("TEST:", test)
    test_dir = join(read_root, test)
    # Get sub directories in the test
    sub_dirs = [dir for dir in os.listdir(test_dir) if isdir(join(test_dir, dir))]

    for sub_dir in sub_dirs:
        # Get final folder with images
        pathIn = [elem for elem in os.walk(join(test_dir, sub_dir))][-1][0]

        # Set path to video file
        out_dir = join(write_root,test)
        
        # If folder doesnt exist - create
        if not os.path.exists(out_dir):
            os.makedirs(out_dir)
            
        pathOut = join(out_dir,sub_dir+'.avi')
        print('Start recording:', pathOut)
        
        # Create video
        create_video(pathIn, pathOut)
        print('End recording:', pathOut)
```
